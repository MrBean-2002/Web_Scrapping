```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.investing.com/'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text,'html')
soup
```

```python
body=[i.text for i in soup.find_all('td')]
body
```

```python
titles=soup.find_all('tr')[0]
titles=[i.text for i in titles.find_all('span')]
titles
```

```python
table=[body[i:i+7] for i in range(0,len(body),7)][:10]
table
```

```python
json=[]
for row in table:
    dictionary={titles[i]:row[i] for i in range(len(row))}
    json.append(dictionary)
json
```
