```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://awsregion.info/'
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
soup
```

```python
scrap=[s.text.strip().split('\n') for s in soup.find_all('tr')]
scrap
```

```python
titles=scrap[0]
```

```python
json=[]
for sub_scrap in scrap[1:-6]:
    my_data={titles[i]: sub_scrap[i] for i in range(len(titles))}
    json.append(my_data)

json         
```
